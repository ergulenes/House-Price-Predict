```python
import pandas as pd
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
import tensorflow
from tensorflow.keras import Input, Model, layers
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt
import glob
```

```python
inputimages = [] #resimleri içine almak için input images listesi tanımlıyoruz
outputimages = np.zeros((64,64,3), dtype="uint8") #sıfırlarla dolu 64,64,3'lük array tanımlıyoruz
```

```python
for housepath in glob.glob('/content/drive/MyDrive/Houses/*.jpg'):
  image = cv2.imread(housepath) #verileri okuyoruz
  image = cv2.resize(image,(32,32),cv2.INTER_AREA) #verilerin pixellerini resize ediyoruz
  inputimages.append(image) #verilerimizi input images'e ekliyoruz
```

```python
four_images = [] #resimleri dörder dörder almak için boş bir liste oluşturuyorum
```

```python
length_images=len(inputimages)
for i in range(0,length_images,4): #her dört resmi bir resim haline getiriyorum.
  outputimages[0:32, 0:32] = inputimages[i]
  outputimages[0:32, 32:64] = inputimages[i+1]
  outputimages[32:64, 32:64] = inputimages[i+2]
  outputimages[32:64, 0:32] = inputimages[i+3]
  four_images.append(outputimages) #resimleri four_images'e ekliyoruz.
```

```python
price = pd.read_csv("/content/drive/MyDrive/Houses Dataset/HousesInfo-2.txt", header = None, sep=" ") #okurken boşluklara göre ayırıyorum ve sütün başlığını kaldırıyorum
price = price.iloc[:,-1:] #price verisini çekiyorum sadece
```

```python
x_train,x_test,y_train,y_test = train_test_split(four_images ,price , test_size=0.33)
```

```python
y_test = y_test.values
y_train = y_train.values
```

```python
x_test = np.array(x_test)
x_train = np.array(x_train)
```

##Model

```python
input_shapes = (64,64,3) 
input_layer = Input(shape=input_shapes)
```

```python
l1 = Conv2D(filters=50, kernel_size=3, strides=(1,1), activation="relu")(input_layer)
l2 = Conv2D(filters=200, kernel_size=3, strides=(1,1), activation="relu")(l1)
max_p1 = MaxPooling2D(pool_size=(2, 2), strides=2, padding="valid")(l2)
l3 = Conv2D(filters=300, kernel_size=3, strides=(1,1), activation="relu")(max_p1)
max_p2 = MaxPooling2D(pool_size=(2, 2), strides=2, padding="valid")(l3)
l4 = Conv2D(filters=500, kernel_size=3, strides=(1,1), activation="relu")(max_p2)
flatten_layer = Flatten()(l4)
dense_layer1 = Dense(3456, activation="relu")(flatten_layer)
dense_layer2 = Dense(1020, activation="relu")(dense_layer1)
dense_layer3 = Dense(200, activation="relu")(dense_layer2)
dense_layer4 = Dense(104, activation="relu")(dense_layer3)
dense_layer5 = Dense(27, activation="relu")(dense_layer4)
output_layer = Dense(1, activation="linear")(dense_layer5)
model = Model(input_layer,output_layer)

"""tensorflow.keras.utils.plot_model(
    model,
    to_file='model.png',
    show_shapes=True,
    show_dtype=True,
    show_layer_names=True,
    rankdir='TB',
    expand_nested=True,
    dpi=96,
    layer_range=None,
    show_layer_activations=True
)"""
model.compile(optimizer="adam",
              loss="mean_absolute_error",
              metrics="mean_absolute_error")
```

```python
model.fit(x = x_train,
          y = y_train,
          batch_size=10,
#         validation_data=(x_test,y_test),
#          verbose=,
          epochs=10)
```
